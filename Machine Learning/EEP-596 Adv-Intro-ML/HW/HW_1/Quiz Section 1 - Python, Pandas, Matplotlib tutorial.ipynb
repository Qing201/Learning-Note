# Processing on Dataframes

## One-hot encoding on categorical variables, make sure consistent between trian & test!

```python
neighborhood = df['Neighborhood']
print(neighborhood)
```

```python
one_hot_neighbor = pd.get_dummies(neighborhood)
print(one_hot_neighbor)
```

## Normalization On Specific Numeric Columns

```python
df_Lot = df[["LotFrontage", "LotArea"]]
print(df_Lot)
```

```python
df_mean_column = df_Lot.mean(axis=0)
df_max_column = df_Lot.max(axis=0)
df_min_column = df_Lot.min(axis=0)
print(df_mean_column, df_max_column, df_min_column)
```

```python
df_Lot_norm = (df_Lot - df_mean_column) / (df_max_column - df_min_column)
print("Numerical Vars after feature scaling(data normalization)!")
print(df_Lot_norm)
```

## Grouping in Dataframe

```python
neighborhood_saleprice = df[['Neighborhood', 'SalePrice']]
price_by_neighborhood = neighborhood_saleprice.groupby('Neighborhood')['SalePrice'].mean()
print(price_by_neighborhood)
```
